```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('placed.csv')
```

```python
X = df.iloc[:,0:1]
y = df.iloc[:,-1]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
y_pred = lr.predict(X_test)
```

```python
# corresponding values of the test set
y_test.values
```

```python
print('MAE', mean_absolute_error(y_test,y_pred))      # MAE in LPA
```

```python
print('MSE', mean_squared_error(y_test,y_pred))
```

```python
print('RMSE', np.sqrt(mean_squared_error(y_test,y_pred)))            # RSME in LPA
```

```python
print('r2_score', r2_score(y_test,y_pred))
```

```python
r2 = r2_score(y_test,y_pred)
```

```python
X_test.shape
```

```python
# Adjusted R2 Score
adj_r2 = 1-((1-r2)*(40-1)/(40-1-1))
adj_r2
```
